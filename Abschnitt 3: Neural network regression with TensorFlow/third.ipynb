```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import pandas as pd
from tensorflow import keras as ks
import os
import logging
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'
logging.getLogger('absl').setLevel(logging.ERROR)
```

```python
x_original = tf.range(start=-7, limit=-7 + 3 * 100, delta=3, dtype=tf.int32)
y_original = tf.range(start=3, limit=3 + 3 * 100, delta=3, dtype=tf.int32)


# x_original = np.array([-7.0, -4.0, -1.0, 2.0, 5.0, 8.0, 11.0, 14.0])
# y_original = np.array([3.0, 6.0, 9.0, 12.0, 15.0, 18.0, 21.0, 24.0])
# x = tf.cast(tf.constant(x_original), dtype=tf.float32)
# y = tf.cast(tf.constant(y_original), dtype=tf.float32)
```

```python
# x_original = tf.reshape(x_original, shape=(100, 1))
# y_original = tf.reshape(y_original, shape=(100, 1))
# x_original, y_original
```

```python
x = tf.identity(x_original)
y = tf.identity(y_original)
```

```python
plt.figure(figsize=(24,12))
plt.scatter(x=x, y=y, c='green', label='original data')
plt.legend()
plt.show()
```

```python
div = int(len(x)*0.7)

x_train = x[0:div]
y_train = y[0:div]

x_test = x[div:]
y_test = y[div:]

div, x_train, y_train, x_test, y_test
```

```python
plt.figure(figsize=(24,12))
plt.scatter(x=x_train, y=y_train, c='blue', label='train data')
plt.scatter(x=x_test, y=y_test, c='green', label='test data')
plt.legend()
plt.show()
```

```python
x = tf.reshape(x, (-1, 1))
y = tf.reshape(y, (-1, 1))

x_train = tf.reshape(x_train, (-1, 1))
y_train = tf.reshape(y_train, (-1, 1))

x_test = tf.reshape(x_test, (-1, 1))
y_test = tf.reshape(y_test, (-1, 1))


x, y
```

```python
x[0].shape, y[0].shape
```

```python
tf.random.set_seed(13)

model1 = tf.keras.Sequential([
    tf.keras.layers.Dense(1),
])

model1.compile(loss=tf.keras.losses.mae,
               optimizer=tf.keras.optimizers.SGD(),
               metrics=['mae'])

model1.fit(x_train,y_train,epochs=100)
```

```python
y_predicts = model1.predict(x_test)
y_predicts
```

```python
y1_predicts = model1.predict(x_test)

plt.figure(figsize=(24,12))
plt.scatter(x=x_train, y=y_train, c='blue', label='train data')
plt.scatter(x=x_test, y=y_test, c='green', label='test data')
plt.scatter(x=x_test, y=y_predicts, c='red', label='predict data')
plt.legend()
plt.show()
```

```python
tf.random.set_seed(13)

model2 = ks.Sequential()
model2.add(ks.layers.Input(1,))
model2.add(ks.layers.Dense(11))
model2.add(ks.layers.Dense(1))

model2.compile(loss=tf.keras.losses.mae,
               optimizer=tf.keras.optimizers.SGD(),
               metrics=['mae'])

model2.fit(x_train,y_train,epochs=100)
```

```python
y2_predicts = model2.predict(x_test)
x_test, y_predicts
```

```python
plt.figure(figsize=(24,12))
plt.scatter(x=x_train, y=y_train, c='blue', label='train data')
plt.scatter(x=x_test, y=y_test, c='green', label='test data')
plt.scatter(x=x_test, y=y2_predicts, c='red', label='predict data')
plt.legend()
plt.show()
```

```python
tf.random.set_seed(42)

model4 = tf.keras.Sequential([
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1),
])

model4.compile(loss=tf.keras.losses.mae,
               optimizer=tf.keras.optimizers.SGD(learning_rate=0.00001),
               metrics=['mae'])

model4.fit(x_train,y_train,epochs=100)

y4_predicts = model4.predict(x_test)

plt.figure(figsize=(24,12))
plt.scatter(x=x_train, y=y_train, c='blue', label='train data')
plt.scatter(x=x_test, y=y_test, c='green', label='test data')
plt.scatter(x=x_test, y=y1_predicts, c='red', label='predict1 data')
plt.scatter(x=x_test, y=y2_predicts, c='yellow', label='predict2 data')
plt.scatter(x=x_test, y=y4_predicts, c='purple', label='predict4 data')
plt.legend()
plt.show()
```
